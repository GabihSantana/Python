<h1>Manipulação de Dados e Criação de Gráficos com bibliotecas Python</h1>

```python
# Pandas -> para manipular dados
# Plotly -> Gráficos

import pandas as pd
import plotly.express as px
```

```python
# Lendo a base de dados do Excel, na folha principal (tem + de 1 folha)
df_principal = pd.read_excel("/content/Imersão Alura.xlsx", sheet_name="Principal")
```

```python
# Lendo as 10 primeiras informações:
df_principal.head(10)
```

```python
# Importanto outra folha, agora a de total de ações
df_total_de_acoes = pd.read_excel("/content/Imersão Alura.xlsx", sheet_name="total_de_acoes")
```

```python
df_total_de_acoes.head(10)
```

```python
# Importanto outra folha, agora a Ticker
df_Ticker = pd.read_excel("/content/Imersão Alura.xlsx", sheet_name="Ticker")
```

```python
df_Ticker.head(10)
```

```python
# Importanto outra folha, agora a que foi criada com ajuda do chatGPT
df_Chat_GPT = pd.read_excel("/content/Imersão Alura.xlsx", sheet_name="Chat_GPT")
```

```python
df_Chat_GPT.head(10)
```

```python
# Atribuindo a df_principal a cópia das colunas que serão utilizadas para fazer os cálculos
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
```

```python
df_principal
```

```python
# Renomeando as colúnas (tirando acentos, pontos e espaços)
df_principal = df_principal.rename(columns={'Último (R$)' : 'valor_final', 'Var. Dia (%)': 'var_dia_pct'}).copy()
```

```python
df_principal
```

```python
# Criando a coluna var_pct (variação %) e fazendo o cálculo
df_principal['var_pct'] = df_principal['var_dia_pct'] / 100
```

```python
df_principal
```

```python
# Criando a coluna valor_inicial e fazendo o cálculo para descobrir o valor inicial
df_principal['valor_inicial'] = df_principal['valor_final'] / ( df_principal['var_pct'] + 1)
```

```python
df_principal
```

VLOOKUP || PROCV <br>
merge -> a outra tabela, cada uma tem um lado, uma fica na esquerda outra na direita e terão suas colunas que são iguais; how: acrescentar os valores da direita na esquerda

Se na coluna principal tem ações que não tem no df_total, retorna NULL, pois mantém os valores chaves e acrescenta a outra coluna

```python
df_principal = df_principal.merge(df_total_de_acoes, left_on='Ativo', right_on='Código', how='left')
```

```python
df_principal
```

```python
# Dropando as colunas repetidas
df_principal = df_principal.drop(columns=['Código'])
```

```python
df_principal
```

```python
# Calculando a variável R$ (diferença do final p inicial dividido pela qtde teórica)
df_principal['var_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial']) * df_principal['Qtde. Teórica']
```

```python
df_principal
```

```python
# Formatando para o formato float e tirando a notação científica
pd.options.display.float_format = '{:.2f}'.format
```

```python
# Modificando o tipo de Qtde Teórica para INT
df_principal['Qtde. Teórica_x'] = df_principal['Qtde. Teórica'].astype(int)
```

```python
df_principal
```

```python
# Renomeando a coluna
df_principal = df_principal.rename(columns={'Qtde. Teórica' : 'Qtde_teorica'}).copy()
```

```python
df_principal
```

Apply -> Função chamada para aplicar outra função - nesse caso, irá atribuir valores de acordo com as condicionais <br>
Lambda -> Usada para operações linha a linha (Operação que deverá ser feito em cada linha) <br>
Em excel: =IF(S2>0,"Subiu", if(S2 < 0, "Desceu", "Estável"))

```python
df_principal['Resultado'] = df_principal['var_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
```

```python
df_principal
```

```python
#VLOOKUP ou PROCV
df_principal = df_principal.merge(df_Ticker, left_on='Ativo', right_on='Ticker', how='left')
```

```python
df_principal
```

```python
# Dropando as colunas repetidas dps do merge
df_principal = df_principal.drop(columns='Ticker')
```

```python
df_principal
```

```python
# Renomeando
df_principal = df_principal.rename(columns={'Nome' : 'nome_empresa'})
```

```python
df_principal
```

```python
df_principal = df_principal.merge(df_Chat_GPT, left_on='nome_empresa', right_on='Empresa', how='left')
```

```python
df_principal
```

```python
df_principal = df_principal.drop(columns='Empresa')
```

```python
df_principal
```

```python
# Criando a categoria de idade e aplicando a condicional em todas as células
df_principal['cat_idade'] = df_principal['Idade (anos)'].apply(lambda x: 'Mais de 100 anos' if x > 100 else ('Menos de 50' if x < 50 else 'Entre 50 e 100'))
```

```python
df_principal
```

Análises

```python
# Cálculo do Maior, menor valor e da média da variação
maior = df_principal['var_rs'].max()
menor = df_principal['var_rs'].min()
media = df_principal['var_rs'].mean()

# Média de quem subiu e de quem desceu
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['var_rs'].mean()
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['var_rs'].mean()

# Printando já com a formatação de float
print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Média\tR$ {media:,.2f}")

print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu:,.2f}")

```

```python
# Atibuindo a df_principal_subiu apenas aqueles que subiram *como filtro*
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

reset_index: Muda a forma de como será printado - Como vetor ou tabela

```python
# Filtro para pegar o segmento -> agrupar pelo Segmento e a soma das variaveis e utilizando .reset_index() p modificar a view
df_analise_segmento = df_principal_subiu.groupby('Segmento')['var_rs'].sum().reset_index()
df_analise_segmento
```

```python
# Filtrando a visualização e somando as categorias
df_analise_saldo = df_principal.groupby('Resultado')['var_rs'].sum().reset_index()
df_analise_saldo
```

GRÁFICOS

```python
# Gráfico de Barra para a análise de saldo (quem subiu, desceu ou ficou estável)
fig = px.bar(df_analise_saldo, x='Resultado', y='var_rs', text = 'var_rs', title='Variação Reais por Resultado', color_discrete_sequence=['purple'])
fig.show()
```

```python
fig.update_traces(texttemplate='%{text:,.2f}', textposition='outside')
```

```python
graf_piz = px.pie(df_analise_segmento, names='Segmento', values='var_rs', title='Variação de quem subiu vs. Segmento',
                  color_discrete_sequence=px.colors.sequential.RdBu)
graf_piz.show()
```

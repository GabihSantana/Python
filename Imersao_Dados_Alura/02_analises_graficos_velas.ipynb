```python
# mplfinance: biblioteca especializada em visualização de dados financeiros oferece recursos específicos para criar gráficos de velas, gráficos de linha e outros tipos de gráficos comuns em análise técnica de mercado.
import mplfinance as mpf
# yfinance: biblioteca que permite acessar e baixar dados financeiros diretamente do Yahoo Finance
import yfinance as yf
# plotly.graph_objects: biblioteca de visualização interativa que permite criar gráficos interativos e personalizáveis.
import plotly.graph_objects as go
# plotly.subplots: usada para criar vários gráficos em uma única figura
from plotly.subplots import make_subplots
```

```python
# Baixando os dados do yahoo finance da Petrobrás com a data de vigência
#dados carregados de 1 ano, porém só contabiliza dias úteis
dados = yf.download('PETR4.SA', start='2023-01-01', end='2023-12-31')
dados
```

```python
# Modificando o nome de TODAS as colunas:
dados.columns = ['Abertura', 'Maximo', 'Minimo', 'Fechamento', 'Fech_Ajust', 'Volume']
```

```python
dados
```

```python
# Renomeando a axis (data) que não foi modificada anteriormente
dados = dados.rename_axis('Data')
dados
```

```python
# Plotando a linha de fechamento em relação aos meses / ano
dados['Fechamento'].plot(figsize=(10,6))
# Nomeando o plot
plt.title('Variação do Preço por Data', fontsize=16)
#Legenda da linha
plt.legend(['Fechamento'])
```

```python
# Pegando apenas os 60 primeiros dados
df = dados.head(60).copy()

# Convertendo o índice em uma coluna de data
df['Data'] = df.index

# Convertendo as datas para o formato númerico de matplotlib para plotar corretamente os dados nos gráficos

# Apply -> Função q recebe outra função para modificar todas séries de datas
# mdates -> manipular datas
df['Data'] = df['Data'].apply(mdates.date2num)
df
```

```python
#plt.subplots() é uma função que retorna uma figura (fig) e um conjunto de eixos (ax)
fig, ax = plt.subplots(figsize=(15, 8))

# Definir a largura dos candles no gráfico:
width = 0.7

# len -> pega o tamanho do df (60), range -> de 0 ao tamanho de len(df) = 59
for i in range(len(df)):

    # Determina a cor do candlestick
    # Se o preço de fechamento for maior que o de abertura, o candle é verde (a ação valorizou nesse dia).
    # Se for menor, o candle é vermelho (a ação desvalorizou).

  #iloc -> é um método do Pandas que permite acessar os dados de um DataFrame usando a posição numérica das linhas e/ou colunas
    if df['Fechamento'].iloc[i] > df['Abertura'].iloc[i]:
        color = 'green'
    else:
        color = 'red'


    # Desenhando a linha vertical do candle (mecha)
    # Essa linha mostra os preços máximo (topo da linha) e mínimo (base da linha) do dia.
    # Usamos `ax.plot` para desenhar uma linha vertical.
    # [df['Data'].iloc[i], df['Data'].iloc[i]] define o ponto x da linha (a data), e [df['Mínimo'].iloc[i], df['Máximo'].iloc[i]] define a altura da linha.

    ax.plot([df['Data'].iloc[i], df['Data'].iloc[i]], # A linha será traçada no mesmo dia
            [df['Minimo'].iloc[i], df['Maximo'].iloc[i]], # Define os pontos min e max
            color=color, # A cor será a mesma definida anteriormente
            linewidth=1)

    # patch -> add um objeto em ax
    # plt.Rectangle -> adiciona o retângulo
    # abs -> é como se fosse um módulo em mat (valor smp positivo)
    ax.add_patch(plt.Rectangle((df['Data'].iloc[i] - width/2, min(df['Abertura'].iloc[i], df['Fechamento'].iloc[i])),
                               width,
                               abs(df['Fechamento'].iloc[i] - df['Abertura'].iloc[i]),
                               facecolor=color))

#                 MEDIA MÓVEL

#rolling -> rola no data frame de 7 em 7 e gera a média desses 7 pontos até acabar o df
df['MA7'] = df['Fechamento'].rolling(window=7).mean()
#rolling -> rola no data frame de 14 em 14 e gera a média desses 14 pontos até acabar o df
df['MA14'] = df['Fechamento'].rolling(window=14).mean()

# Plotando as médias móveis
ax.plot(df['Data'], df['MA7'], color='pink', label='Média Móvel 7 Dias')  # Média de 7 dias
ax.plot(df['Data'], df['MA14'], color='orange', label='Média Móvel 14 Dias')  # Média de 14 dias

# Adicionando legendas para as médias móveis
ax.legend()

# Formatando o eixo x para mostrar as datas
# Configuramos o formato da data e a rotação para melhor legibilidade
ax.xaxis_date() # O método xaxis_date() é usado para dizer ao Matplotlib que as datas estão sendo usadas no eixo x
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d')) # Formatando a data
plt.xticks(rotation=45) # Rotação da data no eixo X

# Adicionando título e rótulos para os eixos x e y
plt.title("Gráfico de Candlestick - PETR4.SA com matplotlib")
plt.xlabel("Data")
plt.ylabel("Preço")

# Adicionando uma grade para facilitar a visualização dos valores
plt.grid(True)

# Exibindo o gráfico
plt.show()
```

*** CRIANDO *SUBPLOTS***

```python
'''
Primeiro, criamos uma figura que conterá os gráficos usando ** make_subplots **
> Isso nos permite ter múltiplos gráficos em uma única visualização.
Aqui, teremos ** dois subplots: um para o gráfico de candlestick e outro para o volume de transações. **

'''

# Criando 2 gráficos:
# Rows = 2 gráficos (2 linhas), que vão compartilhar o eixo X
fig = make_subplots(rows=2, cols=1, shared_xaxes=True,
                    vertical_spacing=0.1, #GAP between them
                    subplot_titles=('Candlesticks', 'Volume Transacionado'),
                    row_width=[0.2, 0.7])

'''
No gráfico de candlestick, cada candle representa um dia de negociação,
mostrando o preço de abertura, fechamento, máximo e mínimo. Vamos adicionar este gráfico à nossa figura.
'''
# Adicionando o gráfico de candlestick
# x -> todas as datas
# go.Candlestick -> Já cria o gráfico candle
fig.add_trace(go.Candlestick(x = df.index,
                             open = df['Abertura'],
                             high = df['Maximo'],
                             low = df['Minimo'],
                             close = df['Fechamento'],
                             name = 'Candlestick'),
                             row=1, col=1)

# Adicionamos também médias móveis ao mesmo subplot para análise de tendências
# go.Scatter é um tipo de trace (traço) no Plotly que cria um gráfico de dispersão.
fig.add_trace(go.Scatter(x = df.index,
                         y = df['MA7'],
                         mode = 'lines',
                         name = 'MA7 - Média Móvel 7 Dias'),
                         row=1, col=1)

fig.add_trace(go.Scatter(x = df.index,
                         y = df['MA14'],
                         mode = 'lines',
                         name = 'MA14 - Média Móvel 14 Dias'),
                         row=1, col=1)


# Adicionando o gráfico de barras para o volume
# Em seguida, criamos um gráfico de barras para o volume de transações, que nos dá uma ideia da atividade de negociação naquele dia
fig.add_trace(go.Bar(x = df.index,
                     y = df['Volume'],
                     name = 'Volume'),
                     row=2, col=1)

# Atualizando layout
# Finalmente, configuramos o layout da figura, ajustando títulos, formatos de eixo e outras configurações para tornar o gráfico claro e legível.
fig.update_layout(yaxis_title='Preço',
                  xaxis_rangeslider_visible=False,  # Desativa o range slider - Não dá para mexer para o lado (ver mais dados)
                  width=1100, height=600)

# Mostrando o gráfico
fig.show()
```

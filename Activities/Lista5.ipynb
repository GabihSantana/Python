```python
media_gov = dados['confian_gov'].mean()
```

```python
print(f"Taxa de confinça no governo é de {media_gov:.2f}")
```

```python
def rotulo_confianca(valor):
    if valor >= media_gov:
        return "Alta"
    else:
        return "Baixa"
```

```python
dados['Rotulo_Confianca'] = dados['confian_gov'].apply(rotulo_confianca)
```

```python
print(dados.head(5))
```

```python
media_por_regiao = dados.groupby('Rotulo_Confianca')['Pontuacao_Felicidade'].mean()
```

```python
print(media_por_regiao)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
media_por_regiao.plot(kind='bar', color='lightpink') ## bar = grafico em barras  ## lightpink é a cor do grafico
plt.title('Confiança e Felicidade entre os Países')  ## dê um titulo 
plt.xlabel('Nível de confiança no governo')  ## o que você quer no eixo x
plt.ylabel('Média de felicidade')    ## o que você quer no eixo y
plt.xticks(rotation=0) ##rotação do título no eixo x
plt.show()
```

```python
# Defina os dados para os eixos x e y
x = dados['PIB_per_capita']
y = dados['Pontuacao_Felicidade']

# Crie o gráfico de dispersão
plt.figure(figsize=(10, 6))  # Define o tamanho da figura
plt.scatter(x, y, alpha=0.5)  # Cria o scatter plot com transparência

# Adicione rótulos aos eixos
plt.xlabel('PIB per capita')
plt.ylabel('Pontuação de Felicidade')
plt.title('Relação do PIB com Felicidade em 2019')

# Exiba o gráfico
plt.show()
```

```python
media_felicidade_total = dados['Pontuacao_Felicidade'].mean()
```

```python
# Crie DataFrames separados para alta e baixa felicidade
acima_media = dados[dados['Pontuacao_Felicidade'] >= media_felicidade_total]
baixa_media = dados[dados['Pontuacao_Felicidade'] < media_felicidade_total]
```

```python
acima_media
```

```python
baixa_media
```

```python
contagem_valores = acima_media['Continentes'].value_counts()
```

```python
contagem_valores_b = baixa_media['Continentes'].value_counts()
```

```python
contagem_valores
```

```python
contagem_valores_b
```

```python
porcentagem_acima_da_media = (contagem_valores / len(acima_media)) * 100
```

```python
porcentagem_acima_da_media
```

```python
porcentagem_abaixo_da_media = (contagem_valores_b / len(baixa_media)) * 100
```

```python
porcentagem_abaixo_da_media
```

```python
import matplotlib.pyplot as plt

# Gráfico de barras para o grupo de alta felicidade

plt.figure(figsize=(10, 6))
plt.bar(porcentagem_acima_da_media.index, porcentagem_acima_da_media)
plt.title('Repartição da felicidade por Continente - Alta Felicidade (2019)')
plt.xlabel('Continente')
plt.ylabel('Porcentagem')
plt.xticks(rotation=45)

# Exibir o gráfico
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(porcentagem_abaixo_da_media.index, porcentagem_abaixo_da_media)
plt.title('Repartição da felicidade por Continente - Baixa Felicidade (2019)')
plt.xlabel('Continente')
plt.ylabel('Porcentagem')
plt.xticks(rotation=45)

# Exibir o gráfico
plt.show()
```
